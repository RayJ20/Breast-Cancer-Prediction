```python
from sklearn.datasets import load_breast_cancer
ds=load_breast_cancer()
```

```python
ds
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
X=pd.DataFrame(ds['data'],columns=ds['feature_names'])
```

```python
X.head()
```

```python
y=pd.DataFrame(data=ds['target'])
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
params=[{'C':[1,5,10]},{'max_iter':[100,150]}]
```

```python
from sklearn.model_selection import GridSearchCV
model1=GridSearchCV(model,params,scoring='f1',cv=10)
model1.fit(X_train,y_train)
```

```python
print(model1.best_params_)
```

```python
model1.best_score_
```

```python
y_pred=model1.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
score=accuracy_score(y_pred,y_test)
```

```python
score
```

```python
confusion_matrix(y_pred,y_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_test,X_train,y_test,y_train=train_test_split(X,y,test_size=0.33,random_state=42)
scalar=StandardScaler()
```

```python
X_test=scalar.fit_transform(X_test)
X_train=scalar.transform(X_train)
y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
error_rate=[]
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred=knn.predict(X_test)
    error_rate.append(pred)
```

```python
from sklearn.metrics import accuracy_score
k=[]
for i in range(0,39):
    score=accuracy_score(error_rate[i],y_test)
    k.append(score)
```

```python
a=max(k)
print(k.index(a))
print(a)
```
